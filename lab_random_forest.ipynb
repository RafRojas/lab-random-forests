### Load data:

```python
numerical_df = pd.read_csv('files_for_lab/numerical.csv')
categorical_df = pd.read_csv('files_for_lab/categorical.csv')
target_df = pd.read_csv('files_for_lab/target.csv')
```

```python
df = round(pd.concat([numerical_df, categorical_df, target_df], axis=1),2)

df.head()
```

### Upsampling:

```python
data = df.copy()   # Creating a copy just in case.

length_0 = len(data[data['TARGET_B']==0]) 
length_1 = len(data[data['TARGET_B']==1])

# Using random sampling to even out when TARGET_B is both equal to 0 and 1.

target_b1 = data[data['TARGET_B']==1].sample(length_0, replace = True) 
target_b0 = data[data['TARGET_B']==0]

print(target_b1.shape)
print(target_b0.shape)
```

```python
data1 = df.copy()
from sklearn.utils import resample

no_donate = data1[data1['TARGET_B']==0] 
yes_donate = data1[data1['TARGET_B']==1]

yes_upsample = resample(yes_donate,
                       replace=True,
                       n_samples = len(no_donate),
                       random_state=23)

display(yes_upsample.shape)
display(no_donate.shape)
```

```python
upsample = pd.concat([yes_upsample, no_donate])
```

```python
df_upscaled = pd.concat([target_b1, target_b0], axis = 0)

df_upscaled = df_upscaled.sample(frac=1).reset_index(drop=True)

df_upscaled
```

### X and y split:

```python
y = upsample['TARGET_B']
X = upsample.drop(['TARGET_B'], axis=1)
```

```python
from sklearn.model_selection import train_test_split

X_train, X_test, y_train, y_test = train_test_split(X, y, test_size = 0.3, random_state = 42)
```

```python
X_train_numerical = X_train.select_dtypes(np.number)
X_train_categorical = X_train.select_dtypes(object)

X_test_numerical = X_test.select_dtypes(np.number)
X_test_categorical = X_test.select_dtypes(object)
```

### Encoding categoricals:

```python
from sklearn.preprocessing import OneHotEncoder

encoder = OneHotEncoder()
encoder.fit(X_train_categorical)

X_train_cat_enc = encoder.transform(X_train_categorical).toarray()
X_train_cat_enc = pd.DataFrame(X_train_cat_enc, columns=encoder.get_feature_names_out())

X_test_cat_enc = encoder.transform(X_test_categorical).toarray()
X_test_cat_enc = pd.DataFrame(X_test_cat_enc, columns=encoder.get_feature_names_out())

X_train_cat_enc.head()
```

### Standardizing numericals:

```python
from sklearn.preprocessing import StandardScaler

scaler = StandardScaler()
scaler.fit(X_train_numerical)

X_train_num = scaler.transform(X_train_numerical)
X_train_num = pd.DataFrame(X_train_num, columns=X_train_numerical.columns)
X_train_num.head()
```

```python
X_test_treated = scaler.transform(X_test_numerical)
X_test_num = pd.DataFrame(X_test_treated, columns=X_test_numerical.columns)
X_test_num.head()
```

### Concatenating for the final X_train and X_test:

```python
train_final = pd.concat([X_train_num, X_train_cat_enc], axis=1)
test_final = pd.concat([X_test_num, X_test_cat_enc], axis=1)

#X_train_final.head()
print(train_final.shape)
print(test_final.shape)
```

### Random forest:

```python
from sklearn.ensemble import RandomForestClassifier
from sklearn.metrics import confusion_matrix
from sklearn.model_selection import cross_val_score

clf = RandomForestClassifier(max_depth=5,
                             min_samples_split=2,
                             min_samples_leaf =4,
                             max_samples=0.5,
                             random_state = 42)

clf.fit(train_final, y_train)
y_pred_train = clf.predict(test_final)
#print("The R2 for the Random Forest in the TRAIN set is {:.2f}".format(clf.score(X_train, y_train)))
#print("The R2 for the Random Forest in the TEST  set is {:.2f}".format(clf.score(X_test, y_test)))
#print("The mean R2 of the folds was {:.2f}".format(np.mean(cross_val_scores)))
```

```python
def model_performance(y_train, y_pred_train, y_test, y_pred_test):

    from sklearn.metrics import mean_absolute_error, mean_squared_error, r2_score

    ME_train = np.mean(y_train-y_pred_train)
    ME_test  = np.mean(y_test-y_pred_test)

    MAE_train = mean_absolute_error(y_train,y_pred_train)
    MAE_test  = mean_absolute_error(y_test,y_pred_test)

    MSE_train = mean_squared_error(y_train,y_pred_train)
    MSE_test  = mean_squared_error(y_test,y_pred_test)

    RMSE_train = np.sqrt(MSE_train)
    RMSE_test  = np.sqrt(MSE_test)

    MAPE_train = np.mean((np.abs(y_train-y_pred_train) / y_train)* 100.)
    MAPE_test  = np.mean((np.abs(y_test-y_pred_test) / y_test)* 100.)

    R2_train = r2_score(y_train,y_pred_train)
    R2_test  = r2_score(y_test,y_pred_test)

    performance = pd.DataFrame({'Error_metric': ['Mean error','Mean absolute error','Mean squared error',
                                             'Root mean squared error','Mean absolute percentual error',
                                             'R2'],
                            'Train': [ME_train, MAE_train, MSE_train, RMSE_train, MAPE_train, R2_train],
                            'Test' : [ME_test, MAE_test , MSE_test, RMSE_test, MAPE_test, R2_test]})

    pd.options.display.float_format = '{:.2f}'.format

    df_train = pd.DataFrame({'Real': y_train.tolist(), 'Predicted': y_pred_train.tolist()})
    df_test  = pd.DataFrame({'Real': y_test.tolist(),  'Predicted': y_pred_test.tolist()})

    return performance, df_train, df_test
```
